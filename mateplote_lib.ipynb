```python
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
```

```python
lst = [4,7,9,6]

plt.plot(lst,marker='*',ms=10,mfc='yellow',color = 'red',ls='dashed') #marker='*',ms=10,mfc='yellow',color = 'red',ls='dashed represent of point 
plt.title('Demo line chart')
plt.ylabel("random number")
plt.xlabel("real data")
plt.show() #skeling value in 2d dimenations 
```
